```python
import json
import sys
sys.path.append('../')
import clustering_tester
```

```python
def load_json(filename):
    with open(filename, 'r', encoding='utf8') as f:
        return json.load(f)
    
```

```python
prefixes = [
  './quanthd_isolet_01_',
  './quanthd_isolet_02_',
  './quanthd_mnist_01_',
  './quanthd_mnist_02_',
]
postfixes = [
  ('noerror', '0.00%'),
  ('27C_Front_10nm_3bit', '0.05%'),
  ('80C_Back_3nm_3bit', '5.22%'),
  ('27C_Front_3nm_3bit', '39.71%'),
  ('27C_Front_10nm_4bit', '7.86%'),
  ('27C_Back_3nm_4bit', '19.09%'),
  ('80C_Back_3nm_4bit', '35.28%'),
]
```

```python
clustering_tester.plot_histories(load_json('./quanthd_isolet_fullprecision.json'), 'isolet full-precision', (0.4, 0.7))
```
